```python
def extract_frames(video_path):
    cap = cv2.VideoCapture(video_path)
    frames = []

    while True:
        ret, frame = cap.read()
        if not ret:
            break
        frames.append(frame)

    return frames
```

```python
import cv2
import numpy as np
import mediapipe as mp

mp_holistic = mp.solutions.holistic

def extract_keypoints_from_video(video_path, max_frames=None):
    cap = cv2.VideoCapture(video_path)

    all_keypoints = []

    with mp_holistic.Holistic(
        static_image_mode=False,
        model_complexity=1,
        enable_segmentation=False,
        refine_face_landmarks=False
    ) as holistic:

        frame_idx = 0

        while True:
            ret, frame = cap.read()
            if not ret:
                break

            if max_frames is not None and frame_idx >= max_frames:
                break

            # OpenCV gives BGR, MediaPipe expects RGB
            rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            results = holistic.process(rgb)

            keypoints = []

            # Pose: 33 landmarks × 4 values = 132
            if results.pose_landmarks:
                for lm in results.pose_landmarks.landmark:
                    keypoints.extend([lm.x, lm.y, lm.z, lm.visibility])
            else:
                keypoints.extend([0] * 33 * 4)

            # Left hand: 21 landmarks × 3 values = 63
            if results.left_hand_landmarks:
                for lm in results.left_hand_landmarks.landmark:
                    keypoints.extend([lm.x, lm.y, lm.z])
            else:
                keypoints.extend([0] * 21 * 3)

            # Right hand: 21 landmarks × 3 values = 63
            if results.right_hand_landmarks:
                for lm in results.right_hand_landmarks.landmark:
                    keypoints.extend([lm.x, lm.y, lm.z])
            else:
                keypoints.extend([0] * 21 * 3)

            all_keypoints.append(keypoints)
            frame_idx += 1

    cap.release()

    return np.array(all_keypoints, dtype=np.float32)
```

```python
from tqdm import tqdm
import os
import numpy as np

ROOT = "/Volumes/T7/Bornil_Dataset/archive"
FEATURE_DIR = "/Volumes/T7/Bornil_Dataset/features_mediapipe"
os.makedirs(FEATURE_DIR, exist_ok=True)

def process_dataset(df, limit=20, max_frames=100):
    for i in tqdm(range(min(limit, len(df)))):
        row = df.iloc[i]

        video_path = os.path.join(ROOT, row["chunk_path"], row["recording"])
        save_name = row["recording"].replace(".webm", ".npy")
        save_path = os.path.join(FEATURE_DIR, save_name)

        if os.path.exists(save_path):
            continue

        if not os.path.exists(video_path):
            print("Missing:", video_path)
            continue

        features = extract_keypoints_from_video(video_path, max_frames=max_frames)
        np.save(save_path, features)

process_dataset(df, limit=20, max_frames=100)
```

```python
from collections import Counter

def clean_text(text):
    text = str(text).strip()
    text = " ".join(text.split())
    return text

df["clean_text"] = df["text"].apply(clean_text)

word_counter = Counter()

for sentence in df["clean_text"]:
    word_counter.update(sentence.split())

special_tokens = ["<pad>", "<bos>", "<eos>", "<unk>"]

word2idx = {tok: i for i, tok in enumerate(special_tokens)}

for word, count in word_counter.items():
    if word not in word2idx:
        word2idx[word] = len(word2idx)

idx2word = {i: w for w, i in word2idx.items()}

print("Vocabulary size:", len(word2idx))
print("Most common words:", word_counter.most_common(20))
```

```python
def encode_sentence(sentence):
    words = clean_text(sentence).split()
    ids = [word2idx["<bos>"]]
    ids += [word2idx.get(w, word2idx["<unk>"]) for w in words]
    ids.append(word2idx["<eos>"])
    return ids

tokens = encode_sentence(df.iloc[0]["text"])

print(df.iloc[0]["text"])
print(tokens)
```

```python
import torch
from torch.utils.data import Dataset, DataLoader
import numpy as np

class SignDataset(Dataset):
    def __init__(self, dataframe):
        self.df = dataframe.reset_index(drop=True)

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        row = self.df.iloc[idx]

        x = np.load(row["feature_path"])          # shape: (T, 258)
        y = row["token_ids"]                      # list of token ids

        x = torch.tensor(x, dtype=torch.float32)
        y = torch.tensor(y, dtype=torch.long)

        return x, y
```

```python
def collate_fn(batch):
    xs, ys = zip(*batch)

    x_lengths = torch.tensor([x.shape[0] for x in xs], dtype=torch.long)
    y_lengths = torch.tensor([y.shape[0] for y in ys], dtype=torch.long)

    xs_padded = torch.nn.utils.rnn.pad_sequence(xs, batch_first=True, padding_value=0.0)
    ys_padded = torch.nn.utils.rnn.pad_sequence(
        ys, batch_first=True, padding_value=word2idx["<pad>"]
    )

    return xs_padded, ys_padded, x_lengths, y_lengths
```

```python
train_dataset = SignDataset(train_df)

train_loader = DataLoader(
    train_dataset,
    batch_size=4,
    shuffle=True,
    collate_fn=collate_fn
)

batch = next(iter(train_loader))

x, y, x_lengths, y_lengths = batch

print(x.shape)
print(y.shape)
print(x_lengths)
print(y_lengths)
```
